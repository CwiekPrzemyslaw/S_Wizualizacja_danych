```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dane = pd.read_csv('dane.csv', sep = ';', decimal = ',',
                   index_col=0)
```

```python
bottom_pos = dane['Rozp_kariery']
```

```python
max_wiek = max(dane['Kon_kariery'])
min_wiek = min(dane['Rozp_kariery'])
```

```python
avg_gs1 = dane['GS1'].mean()
avg_gs2 = dane['GS2'].mean()
avg_gs = (avg_gs2 + avg_gs1) / 2
```

```python
plt.bar(x = dane['Zawodnik'], height = dane['Kon_kariery'] - dane['Rozp_kariery'], bottom = bottom_pos, width=0.1)
plt.ylim(min_wiek - 2, max_wiek + 2)
plt.grid(axis='x', linestyle='-', linewidth=0.5, color='grey')
plt.scatter(dane['Zawodnik'], dane['GS1'], color='red') 
plt.scatter(dane['Zawodnik'], dane['GS2'], color='red')
for i, row in dane.iterrows():
    if not pd.isna(row['GS1']):
        plt.axhline(y=row['GS1'], color='grey', linestyle='--')
    if not pd.isna(row['GS2']):
        plt.axhline(y=row['GS2'], color='grey', linestyle='--')
plt.axhline(y=avg_gs, color='orange', linestyle='-', label='Mean of Average GS1 and GS2')
plt.show()
```
